```python
import torch 
import numpy as np 
import matplotlib.pyplot as plt 
import torch.nn as nn 

x = torch.tensor(np.linspace(-10, 10, 300))
y_relu = nn.ReLU()(x)
y_sigmoid = nn.Sigmoid()(x)
y_leaky_relu = nn.LeakyReLU()(x)
y_tanh = nn.Tanh()(x)

plt.ylim(-1, 1)
plt.xlim(-4, 4)
plt.plot(x, y_relu, label="ReLU") 
plt.plot(x, y_sigmoid, label="Sigmoid") 
plt.plot(x, y_leaky_relu, label="Leaky ReLU") 
plt.plot(x, y_tanh, label="Tanh")
plt.legend() 
```

```python
def swish(x): 
    return x * torch.sigmoid(x) 

def mish(x): 
    return x * torch.tanh(nn.Softplus()(x))

y_swish = swish(x) 
y_mish = mish(x)

plt.ylim(-0.5, 2)
plt.xlim(-4, 4)
plt.plot(x, y_swish, label="Swish") 
plt.plot(x, y_mish, label="Mish") 
plt.legend()
plt.show() 
```

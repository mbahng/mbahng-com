```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression  
```

```python
def true_f(x): 
    return np.sin(2* np.pi * x) + 2 * np.cos(x - 1.5) 

def generate_dataset(num_datasets=10, size=15): 
    datasets = [] 
    for i in range(num_datasets): 
        x = np.random.uniform(0, 2, size=size) 
        y = true_f(x) + np.random.normal(scale=0.3, size=x.shape) 
        datasets.append((x, y)) 
    return datasets
```

```python
Ds = generate_dataset()
```

```python
D = Ds[0] 
X, y = D
plt.scatter(X, y, color='blue', s=5) 
x_plot = np.linspace(0, 2, 100) 
y_plot = true_f(x_plot)
plt.plot(x_plot, y_plot, color="red", linewidth=1)
plt.show() 
plt.clf()
```

```python
for deg in [1, 3, 5, 7, 9, 11]: 
    poly = PolynomialFeatures(degree=deg) 
    for i in range(10): 
        X, y = Ds[i]
        X = np.array(X).reshape(-1, 1) 
        y = np.array(y)
        poly_feature = poly.fit_transform(X) 
        poly_reg_model = LinearRegression() 
        poly_reg_model.fit(poly_feature, y) 
        
        plot_x = np.linspace(0, 2, 100)
        plot_y = poly_reg_model.predict(poly.fit_transform(plot_x.reshape(-1, 1))) 
        # plt.scatter(X, y, color='blue', s=5)
        plt.plot(plot_x, plot_y, color="blue", linewidth=1)
    if deg > 5: 
        plt.ylim(-5, 5)
    x_true = np.linspace(0, 2, 100) 
    y_true = true_f(x_true)
    plt.plot(x_true, y_true, color="red", linewidth=1)
    plt.show() 
    plt.clf()
```

```python
large_D = generate_dataset(10, 100)

poly = PolynomialFeatures(degree=9) 
for i in range(10): 
    X, y = large_D[i]
    X = np.array(X).reshape(-1, 1) 
    y = np.array(y)
    poly_feature = poly.fit_transform(X) 
    poly_reg_model = LinearRegression() 
    poly_reg_model.fit(poly_feature, y) 
    
    plot_x = np.linspace(0, 2, 100)
    plot_y = poly_reg_model.predict(poly.fit_transform(plot_x.reshape(-1, 1))) 
    # plt.scatter(X, y, color='blue', s=5)
    plt.plot(plot_x, plot_y, color="blue", linewidth=1)
x_true = np.linspace(0, 2, 100) 
y_true = true_f(x_true)
plt.plot(x_true, y_true, color="red", linewidth=1)
plt.show() 
plt.clf()

```
